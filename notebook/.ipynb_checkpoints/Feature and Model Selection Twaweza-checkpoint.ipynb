```python
# Format python plot with Latex
SPINE_COLOR = 'gray'
from math import sqrt

def latexify(fig_width=None, fig_height=None, columns=1):
    """Set up matplotlib's RC params for LaTeX plotting.
    Call this before plotting a figure.
    Parameters
    ----------
    fig_width : float, optional, inches
    fig_height : float,  optional, inches
    columns : {1, 2}
    """

    # code adapted from http://www.scipy.org/Cookbook/Matplotlib/LaTeX_Examples

    # Width and max height in inches for IEEE journals taken from
    # computer.org/cms/Computer.org/Journal%20templates/transactions_art_guide.pdf

    assert(columns in [1,2])

    if fig_width is None:
        fig_width = 3.39 if columns==1 else 6.9 # width in inches

    if fig_height is None:
        golden_mean = (sqrt(5)-1.0)/2.0    # Aesthetic ratio
        fig_height = fig_width*golden_mean # height in inches

    MAX_HEIGHT_INCHES = 8.0
    if fig_height > MAX_HEIGHT_INCHES:
        print("WARNING: fig_height too large:" + fig_height + 
              "so will reduce to" + MAX_HEIGHT_INCHES + "inches.")
        fig_height = MAX_HEIGHT_INCHES

    params = {'backend': 'ps',
              'text.latex.preamble': ['\\usepackage{gensymb}'],
              'axes.labelsize': 10, # fontsize for x and y labels (was 10)
              'axes.titlesize': 14,
              'text.fontsize': 12, # was 10
              'legend.fontsize': 10, # was 10
              'xtick.labelsize': 10,
              'ytick.labelsize': 10,
              'text.usetex': True,
              'figure.figsize': [fig_width,fig_height],
              'font.family': 'serif'
            
    }

    matplotlib.rcParams.update(params)
```

## 2.  Model - based feature Selection : Using **Feature Importance**

### 2. 1 Using Random Forest Classifier

```python
# Define a classifier
rf_clf = RandomForestClassifier(max_depth=15,n_estimators=70, min_samples_leaf=50,
                                  min_samples_split=100, random_state=10)
```

```python
data.columns
```

### In this phase the most 8 preveiling features are: Student_age, PTR, PCR, GPLR, BPLR, PTMR, Parentteacherdiscussion and hh_size, so we are going to use these feautures to the next stage of model selection.

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LogisticRegression
from sklearn import svm
from sklearn import tree
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
from sklearn.metrics import classification_report
#from xgboost import XGBClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis as QDA 
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
from sklearn import model_selection
from sklearn.model_selection import train_test_split
from sklearn.metrics import f1_score, accuracy_score
from time import time
```

```python
feature = ['PTR', 'PCR', 'GPLR', 'BPLR', 'PTMR', 'hh_size','Parentteacherdiscussion']
X_train = data[feature]
y_train = data.Dropout
X_val   = data_val[feature]
y_val   = data_val.Dropout
```

```python
def classifiers(X_train,y_train, X_test, y_test, score='f1'):
    knn     =  KNeighborsClassifier()
    gnb     =  GaussianNB()
    log     =  LogisticRegression()
    dTree   =  tree.DecisionTreeClassifier()
    rForest =  RandomForestClassifier()
    #xgboost =  XGBClassifier()
    mlp     =  MLPClassifier()
    adboost = AdaBoostClassifier()
    lda     = LDA(solver='lsqr', shrinkage='auto')
    qda     = QDA()
    
    model_names = ["Nearest Neighbors",
                   "Naive Bayes",
                   "Logistic",
                   "Decision Tree",
                   "Random Forest",
                   #"XgBoost",
                   "MLP",
                   "AdaBoost",
                   "LDA",
                   "QDA"]
    
    #models = [knn, gnb, log, dTree, rForest, xgboost,mlp,adboost,lda,qda]
    models = [knn, gnb, log, dTree, rForest, mlp,adboost,lda,qda]
     
    name= []
    results=[]
    avg_results = []
    f_score_test= []
    acc_test = []
    model_time = []
    for (i,model) in enumerate(models):
      
        skfold = model_selection.StratifiedKFold(n_splits=5)
        start = time()
        cv_results = model_selection.cross_val_score(model, X_train, y_train, cv=skfold, scoring=score)
        end = time()
        results.append(cv_results)
        avg_results.append(cv_results.mean())
        msg = "%s: %f" % (model_names[i], cv_results.mean())
        print(msg)
        
        print("Run test perfomance")
        model.fit(X_train, y_train)
        y_p = model.predict(X_test)
        f_score = f1_score(y_test, y_p)
        f_score_test.append(f_score)
        acc  = accuracy_score(y_test,y_p)
        acc_test.append(acc)
        model_time.append((start - end))
        
    return avg_results, model_names, f_score_test, acc_test, model_time   
```

```python
result, names, f_score_test, acc_test, model_time  = classifiers(X_train,y_train,X_val,y_val)
```

```python
def plotbar(plot_name, names, result,title,ylabel):
    postion = np.arange(len(names))
    plt.bar(postion, result, align='center', color ='g')
    plt.xticks(postion, names, rotation=0)
    plt.ylabel(ylabel)
    plt.title(title)
    plt.savefig('image/%s.pdf' %(plot_name), format='pdf')
    
```

```python
names = ["KNN",
                   "Naive",
                   "Logistic",
                   "DTree",
                   "RForest",
                   #"XgBoost",
                   "MLP",
                   "AdaBoost",
                   "LDA",
                   "QDA"]
    
```

```python
plotbar("crossvalidation_perfomance", names, result, "F-1 Score Performance", "F-Score \n (Higher is better)")
```

```python
plotbar("Test_perfomance", names, f_score_test, "F-1 Score Test Performance", "F-Score \n (Higher is better)")
```

```python
plotbar("Test_perfomance", names, acc_test, "Accuracy  Test Performance", "Accuracy \n (Higher is better)")
```

```python
time = np.array(model_time) * -1
```

```python
plotbar("Time_perfomance", names,time, "Model time Performance", "Time (s)")
```

####  From the above figure three models i.e Nearest Neighours, Decison Tree and Random Forest show best performance. We will therefore consider these three models in the next stage
